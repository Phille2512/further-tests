# Getting Audio Features 

This Notebook is used to get the audio features of all the tracks from the streaming history. The spotipy library will be used to request all the features based on the track id. 

The resulting dataframe is contained in the folder. The notebook runs quite long.

```python
import json
import time
import spotipy
from spotipy.oauth2 import *
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import json
```

```python
client_credentials_manager = SpotifyClientCredentials("8cc8cdb3b7b34b59afa030e2473d3a3d", "b0e8fb486b1e4bfcb68d9fe80deb63c1")
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager, requests_timeout=None)
```

```python
streaming_hist_0 = pd.read_json("endsong_0.json")
streaming_hist_1 = pd.read_json("endsong_1.json")
streaming_hist_2 = pd.read_json("endsong_2.json")
streaming_hist_3 = pd.read_json("endsong_3.json")
streaming_hist_4 = pd.read_json("endsong_4.json")
streaming_hist_5 = pd.read_json("endsong_5.json")
streaming_hist_6 = pd.read_json("endsong_6.json")
streaming_hist_7 = pd.read_json("endsong_7.json")
```

```python
streaming_history_total = pd.concat([streaming_hist_0, streaming_hist_1], axis=0)
streaming_history_total = pd.concat([streaming_history_total, streaming_hist_2], axis =0)
streaming_history_total = pd.concat([streaming_history_total, streaming_hist_3], axis =0)
streaming_history_total = pd.concat([streaming_history_total, streaming_hist_4], axis =0)
streaming_history_total = pd.concat([streaming_history_total, streaming_hist_5], axis =0)
streaming_history_total = pd.concat([streaming_history_total, streaming_hist_6], axis =0)
streaming_history_total = pd.concat([streaming_history_total, streaming_hist_7], axis =0)
```

We're going to reduce the Dataframe to unique values of tracks + artists. 

```python
unique_artists = streaming_history_total[['master_metadata_track_name', 'master_metadata_album_album_name', 'master_metadata_album_artist_name', 'spotify_track_uri']]
```

```python
unique_artists = unique_artists.drop_duplicates()
```

```python
# Split spotify_track_uri to get track id
ids = unique_artists.spotify_track_uri.str.split(":", expand=True).reset_index()
```

```python
# Had an error concerning types thus converting it to string
ids[2] = ids[2].astype(str)
```

```python
#Generating a list of ids
list_of_ids = ids[2].to_list()
```

```python
list_of_ids
```

```python
# Requesting audiofeatures through looping through the list of ids
list_of_features = []

for i in ids[2]: 
    feature = sp.audio_features(i)
    list_of_features.append(feature)


```

```python
# add list_of_ids to the dataframe to have a key to merge the dataframes
unique_artists["id"] = list_of_ids
```

```python
pd.json_normalize(list_of_features)
```

```python
# Create Dataframe based on the audio feature response. 
df_audio_features = pd.DataFrame(pd.json_normalize(list_of_features))
df_audio_features = pd.DataFrame(pd.json_normalize(df_audio_features[0]))
#df_audio_features.head(5)
len(df_audio_features)

```

```python
# merge df's
unique_artist_audio_features = unique_artists.merge(df_audio_features, how='inner', on='id')
```
